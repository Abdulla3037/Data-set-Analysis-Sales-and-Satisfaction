## **Range of Columns that have Numerical Values** ##

```python
import pandas as pd

df = pd.read_csv('Sales_without_NaNs_v1.3.csv')

print("Sales Before")
print("--------------")
sales_bfmin = df['Sales_Before'].min()
print(f'Maximum Range: ', round(sales_bfmin,2))

sales_bfmax = df['Sales_Before'].max()
print(f'Maximum Range: ', round(sales_bfmax,2))
print(end='\n')

print("Sales After")
print("--------------")
sales_aftermin = df['Sales_After'].min()
print(f'Minimum Range: ', round(sales_aftermin,2))

sales_aftermax = df['Sales_After'].max()
print(f'Maximum Range: ', round(sales_aftermax,2))
print(end='\n')
```

```python
print("Customer Satisfaction Before")
print("-----------------------------")
customer_st_bfmin = df['Customer_Satisfaction_Before'].min()
print(f'Maximum Range: ', round(customer_st_bfmin,2))

customer_st_bfmax= df['Customer_Satisfaction_Before'].max()
print(f'Maximum Range: ', round(customer_st_bfmax,2))
print(end='\n')

print("Customer Satisfaction After")
print("-----------------------------")
customer_st_afmin = df['Customer_Satisfaction_After'].min()
print(f'Maximum Range: ', round(customer_st_afmin,2))

customer_st_afmax= df['Customer_Satisfaction_After'].max()
print(f'Maximum Range: ', round(customer_st_afmax,2))
print(end='\n')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
file_path = "Sales_without_NaNs_v1.3.csv"
data = pd.read_csv(file_path)

# Select required columns
stats_data = data[["Sales_Before", "Sales_After",
                   "Customer_Satisfaction_Before", "Customer_Satisfaction_After"]]

# Calculate statistics
stats_summary = pd.DataFrame({
    "Mean": stats_data.mean(),
    "Median": stats_data.median(),
    "Variance": stats_data.var(),
    "Standard Deviation": stats_data.std()
})

# Transpose for plotting
stats_summary = stats_summary.T

# Plot
plt.figure(figsize=(12, 8))
for i in stats_summary.columns:
    plt.plot(stats_summary.index, stats_summary[i], marker="o", label=i)

plt.title("Statistics for Numerical Columns")
plt.xlabel("Statistic Type")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

## **Label Encoding** ##

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red']})

le = LabelEncoder()
data['Color_encoded'] = le.fit_transform(data['Color'])

print(data)
```

## **One-Hot Encoding** ##

```python
import pandas as pd

data = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red']})

# One-hot encoding with integer values
one_hot = pd.get_dummies(data['Color'], dtype=int)
data = pd.concat([data, one_hot], axis=1)

print(data)
```

## **Ordinal Encoding** ##

```python
from sklearn.preprocessing import OrdinalEncoder

data = pd.DataFrame({'Size': ['Small', 'Medium', 'Large', 'Small']})

encoder = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
data['Size_encoded'] = encoder.fit_transform(data[['Size']])

print(data)
```

## **Frequency Encoding** ##

```python
import pandas as pd

# Sample data
data = pd.DataFrame({'Country': ['USA', 'India', 'USA', 'Bangladesh', 'India', 'USA', 'Nepal', 'Bangladesh']})

# Count/frequency encoding
freq_encoding = data['Country'].value_counts().to_dict()
data['Country_Freq'] = data['Country'].map(freq_encoding)

print(data)
```
